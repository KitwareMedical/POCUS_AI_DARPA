```python
import os
import glob

import numpy as np

import itk
from itk import TubeTK as tube
```

```python
old_files = glob.glob("Data_Raw/PTX-ARUNet/*.mha")
new_files = [os.path.basename(x) for x in old_files]
for old_file,new_file in zip(old_files,new_files):
    img = itk.imread(old_file)
    immath = tube.ImageMath.New(img)
    itk.imwrite(immath.GetOutput(), os.path.join("Data","images",new_file), compression=True)
```

# Normalize

```python
images = glob.glob("Data/images/*.mha")

IM = tube.ImageMath[itk.Image[itk.F,3]].New()

for file in images:
    img = itk.imread(file)
    arr = itk.GetArrayViewFromImage(img)
    print(arr.min(), arr.max(), arr.shape)
    if arr.max() > 1:
        IM.SetInput(img)
        IM.IntensityWindow(0,255,0,1)
        img = IM.GetOutput()
        arr = itk.GetArrayViewFromImage(img)
        print("   ", arr.min(), arr.max())
        itk.imwrite(img, file)
```

```python
labels = glob.glob("Data/labels/*.mha")

IM = tube.ImageMath[itk.Image[itk.UC,3]].New()

for file in labels:
    img = itk.imread(file)
    arr = itk.GetArrayFromImage(img).astype(np.uint8)
    n1 = np.count_nonzero(arr==1)
    n2 = np.count_nonzero(arr==2)
    n3 = np.count_nonzero(arr==3)
    print(arr.min(), arr.max(), arr.shape, n1, n2, n3)
    if arr.max() > 2:
        arr = np.where(arr==3,1,arr)
        img = itk.GetImageFromArray(arr)
        arr = itk.GetArrayFromImage(img)
        n1 = np.count_nonzero(arr==1)
        n2 = np.count_nonzero(arr==2)
        n3 = np.count_nonzero(arr==3)
        print("   ", arr.min(), arr.max(), arr.shape, n1, n2, n3)
        itk.imwrite(img, file)
```
