```python
def fix_names(reference_path, files_to_fix):
    videos = sorted(glob(reference_path+"/*.m??"))
    others = sorted(glob(files_to_fix))
    
    for i,orgfilefull in enumerate(others):
        (orgpath,orgfile) = os.path.split(orgfilefull)
        substr = orgfile.split('_')[1]
        print(orgfile)
        basefiles = [basefile for basefile in videos if substr in basefile]
        (basepath, basefile) = os.path.split(basefiles[0])
        newfile = orgpath+'/'+basefile[:6]+orgfile
        print("  ", newfile)
        os.system('cp '+orgfilefull+' '+newfile)  
```

```python
reference_path = "../../Data/VFoldData/BAMC-PTXNoSliding"
files_to_fix = reference_path + "-Annotations-Linear/[iI]*"
fix_names(reference_path, files_to_fix)
reference_path = "../../Data/VFoldData/BAMC-PTXSliding"
files_to_fix = reference_path + "-Annotations-Linear/[iI]*"
fix_names(reference_path, files_to_fix)
```
